```python
def plot_confusion_matrix(cm, classes,normalize,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center", verticalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def eval_model(model, X, y, cv_num):
  scores = cross_val_score(model, X, y, cv=cv_num)
  predicted = cross_val_predict(model, X, y, cv=cv_num)
  cm = confusion_matrix(y, predicted)

  print("cross val scores: {}".format(scores))
  print("variance of cross val scores: {}".format(scores.var()))
  print("classification_report: {}".format(classification_report(y, predicted)))
  plot_confusion_matrix(cm,[0,1],False)
```

Of the naive models, random forest performed best. Even it had a large range in recall score - 0.99 to 0.72. This model does not have good performance.

Next, the number of features is reduced with SelectKBest.

To find the best value of k for the SelectKBest function, values from 2 to 11 are tried with random forest, the best-performing model found previously. The metric used to evaluate each k value is the variance of the cross validation scores of models trained with the k value. The k value with the least variance will be selected as a small variance implies a lower likelihood of overfitting.

```python
k_dict = {}
for k_i in range(2,11):
  selector = SelectKBest(chi2, k=k_i).fit(X_train, y_train)
  X_train_selected = X_train[X_train.columns[selector.get_support()].values]
  rfc_selected = ensemble.RandomForestClassifier(n_estimators=100, n_jobs=-1)
  k_dict[k_i] = cross_val_score(rfc_selected, X_train_selected, y_train, cv=10).var()

k_dict
```



A k value of 5 is selected as it has the lowest cross validation score variance. Then, the data is oversampled to reduce the extreme class imbalancing. In the original dataset, there were were more than five times as many customers who stayed as those who left.

```python
selector = SelectKBest(chi2, k=5).fit(X_train, y_train)
print(str(X_train.columns.values[selector.get_support(indices=True)]))
X_train_selected = X_train[X_train.columns[selector.get_support()].values]

oversample = SMOTE()
X_selected_and_smitten_train, y_smitten_train = oversample.fit_resample(X_train_selected, y_train)
```

This time, the GridSearchCV function is used to pick the model parameters.

```python
parameters = {'n_estimators':[50,75,100,125,150, 175, 200, 225]}
rfc_selected = ensemble.RandomForestClassifier(n_jobs=-1, random_state=2)
clf = GridSearchCV(rfc_selected, parameters, n_jobs=-1)
clf.fit(X_selected_and_smitten_train, y_smitten_train)
clf.best_params_
```

For the random forest model, 100 estimators gave the best performance.

```python
rfc_grid = ensemble.RandomForestClassifier(n_estimators = 100, n_jobs=-1, random_state=2)
eval_model(rfc_grid, X_selected_and_smitten_train, y_smitten_train, 10 )
```

```python
parameters_lr = {'solver':('liblinear','lbfgs','sag', 'saga'),'penalty':('none','l2', 'elasticnet')}
lr_selected = LogisticRegression(random_state=2, max_iter=10000)
clf_lr = GridSearchCV(lr_selected, parameters_lr, n_jobs=-1)
clf_lr.fit(X_selected_and_smitten_train, y_smitten_train)
clf_lr.best_params_
```

For the logistic regression model, using no penalty and the lbfgs solver provided the best performance.

```python
lr_grid = LogisticRegression(solver='lbfgs', penalty = 'none',random_state=2, max_iter=10000)
lr_grid.fit(X_selected_and_smitten_train, y_smitten_train)
eval_model(lr_grid,X_selected_and_smitten_train, y_smitten_train, 10)
print('{} iterations out of {}'.format(lr_grid.n_iter_[0], lr_grid.max_iter))
```

From SelectKBest we know that the most predictive features are 'number vmail messages,' 'total day minutes,' 'total day charge,' 'total eve minutes,' and 'international plan_yes.'

In summary, the random forest model using oversampling to class balance, feature reduction using SelectKBest, and GridSearchCV performed very well, with precision of 0.87 and 0.88 and recall of 0.89 and 0.87 for classes 0 and 1 respectively.
